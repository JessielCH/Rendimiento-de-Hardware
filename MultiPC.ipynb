```python
import socket
import psutil

# Funciones para obtener información del rendimiento
def obtener_rendimiento_cpu_local():
    return psutil.cpu_percent(interval=1)

def obtener_rendimiento_memoria_local():
    return psutil.virtual_memory().percent

def obtener_rendimiento_red_local():
    return psutil.net_io_counters().bytes_sent, psutil.net_io_counters().bytes_recv

def obtener_temperatura_cpu_local():
    try:
        return psutil.sensors_temperatures()['coretemp'][0].current
    except Exception as e:
        print(f"Error al obtener la temperatura local: {e}")
        return None

# Crear un objeto socket TCP/IP
objetoSocket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# Configurar el servidor para escuchar en el puerto 8080 y vincularse a la dirección IP 192.168.100.1
puerto = 8080
direccion_ip_servidor = "0.0.0.0"
objetoSocket.bind((direccion_ip_servidor, puerto))
objetoSocket.listen(1)

print(f"Esperando conexiones en la dirección IP {direccion_ip_servidor}, puerto {puerto}...")

# Aceptar conexiones entrantes
clienteSocket, clienteDireccion = objetoSocket.accept()
print(f"Conexión establecida con {clienteDireccion}")

# Loop principal del servidor
while True:
    # Recibir solicitud del cliente
    solicitud = clienteSocket.recv(1024).decode()
    
    # Procesar solicitud y enviar respuesta
    if solicitud == "obtener_informacion":
        rendimiento_cpu = obtener_rendimiento_cpu_local()
        rendimiento_memoria = obtener_rendimiento_memoria_local()
        rendimiento_red = obtener_rendimiento_red_local()
        temperatura_cpu = obtener_temperatura_cpu_local()

        datos = f"{rendimiento_cpu},{rendimiento_memoria},{rendimiento_red[0]},{rendimiento_red[1]},{temperatura_cpu}"
        clienteSocket.sendall(datos.encode())
    else:
        clienteSocket.sendall("Solicitud no válida".encode())

# Cerrar el socket
objetoSocket.close()


##CODIGOCLIENTE
##import socket

# Crear un objeto socket TCP/IP
#objetoSocket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# Configurar el servidor al que quieres conectarte
#servidor = "192.168.100.1"  # Reemplaza con la dirección IP de tu servidor
#puerto = 8080

# Conectar al servidor
#objetoSocket.connect((servidor, puerto))

# Enviar solicitud al servidor
#solicitud = "obtener_informacion"
#objetoSocket.sendall(solicitud.encode())

# Recibir respuesta del servidor
#respuesta = objetoSocket.recv(1024).decode()
#print("Información recibida del servidor:", respuesta)

# Cerrar el socket
#objetoSocket.close()

```
